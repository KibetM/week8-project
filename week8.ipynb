there are 1385 rows , for columns that contain 1385 nulls they can be dropped since they have no weight on the dataset

```python
new_loc_filter=location_filter.drop(columns=['excess_mortality_cumulative_absolute','excess_mortality_cumulative','excess_mortality','excess_mortality_cumulative_per_million'])
```

```python
new_loc_filter['date']= pd.to_datetime(new_loc_filter['date'])
```

```python
plt.plot(new_loc_filter['date'],new_loc_filter['total_cases'])
plt.xlabel('Date')
plt.ylabel('Total number of covid caases')
plt.title('Total number of cases over time')
plt.tight_layout()
plt.show()
```

```python
plt.plot(new_loc_filter['date'], new_loc_filter['total_deaths'])
plt.xlabel('Date')
plt.ylabel('Total number of deaths')
plt.title('Total number of covid deaths in Kenya over time')
plt.tight_layout()
plt.show()
```

```python
df_clean = df.dropna(subset=['continent', 'new_cases'])
df_clean['continent'] = df_clean['continent'].astype(str)

plt.bar(df_clean['continent'], df_clean['new_cases'], color='blue')
plt.xlabel('Continent')
plt.ylabel('Number of new cases')
plt.title('Number of new cases in comparison to other countries')
plt.tight_layout()
plt.show()
```

```python
df['death_rate']=df['total_deaths']/df['total_cases']
```

```python
plt.plot(new_loc_filter['date'], new_loc_filter['total_vaccinations'])
plt.xlabel('Date')
plt.ylabel('Number of people vaccinated')
plt.title('Number of kenyans vaccinated over time')
plt.tight_layout()
plt.show()
```

```python
# Calculate % vaccinated
df['% vaccinated'] = df['people_fully_vaccinated'] / df['population']

# Drop rows with NaN values in '% vaccinated'
df_clean = df.dropna(subset=['% vaccinated'])

# Optional: sort and keep top N countries (e.g., top 10)
df_top = df_clean.sort_values(by='% vaccinated', ascending=False).head(10)

# Prepare labels and sizes
labels = df_top['location']
sizes = df_top['% vaccinated']

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Top 10 Countries by % Fully Vaccinated')
plt.axis('equal')
plt.show()
```
